```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def read_file(filename):
  img=cv2.imread(filename)
  img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.show()
  return img
```
